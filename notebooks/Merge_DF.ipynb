```python
# check missing dates
def check_missing_dates(df, dateCol):
    subset = df.copy()
    subset = subset.set_index(dateCol)
    subset.index = pd.to_datetime(subset.index)
    print(pd.date_range(
        start="2020-12-14", end="2022-03-01").difference(subset.index))
```

```python
df_final = pd.read_parquet('../data/final_dataset.parquet.gzip')
df_final.head()
```

```python
check_missing_dates(df_final,dateCol='date')
```
