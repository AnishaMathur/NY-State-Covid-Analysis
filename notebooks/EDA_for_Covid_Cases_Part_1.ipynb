```python
#Crude Mortality Rate / Crude Death rate 
df_raw_cmr = df3.filter(['total_deaths','population','population_density','date'], axis=1)
df_cmr = df_raw_cmr.dropna()
df_cmr['CMR'] = (df_cmr['total_deaths']/df_cmr['population'])*100


#Crude Mortality Rate
plt.figure(figsize=(20,4))
plt.plot(df_cmr['date'], df_cmr['CMR'])
plt.xlabel("date", fontsize=14)
plt.ylabel("CMR", fontsize=14)
```

```python
#Infection Fatality Rate 
#?if someone is infected with the disease how likely is it that they will die from it?
#IFR = (number of deaths/total number of cases ) * 100


df_raw_ifr = df3.filter(['total_deaths','total_cases','date'], axis=1)
df_ifr = df_raw_ifr.dropna()
df_ifr['IFR'] = (df_ifr['total_deaths']/df_ifr['total_cases'])*100


plt.figure(figsize=(20,4))
plt.plot(df_ifr['date'], df_ifr['IFR'])
plt.xlabel("date", fontsize=14)
plt.ylabel("IFR", fontsize=14)
```
