```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
path = '../data/'
```

```python
cases = pd.read_parquet(path+'NY/countywise_daily_cases.parquet.gzip')
deaths = pd.read_parquet(path+'NY/countywise_daily_deaths.parquet.gzip')
```

```python
print(cases[cases['daily_cases']<0])
print(deaths[deaths['daily_deaths']<0])
```

```python
print(cases.isna().values.any())
print(deaths.isna().values.any())
```

```python
def plot_graphs(df, county, col):
  temp = df[df['county']==county].reset_index()
  temp['rolling_avg'] = temp[col].rolling(7).mean()
  figure(figsize=(15, 6))
  if col == 'daily_cases':
    label = 'Cases'
  elif col == 'daily_deaths':
    label = 'Deaths'
  plt.plot(temp['date'], temp['rolling_avg'], label=label)
```

```python
plot_graphs(cases, 'Albany', 'daily_cases')
plot_graphs(deaths, 'Albany', 'daily_deaths')
```

```python
plot_graphs(cases, 'Hamilton', 'daily_cases')
plot_graphs(deaths, 'Hamilton', 'daily_deaths')
```
