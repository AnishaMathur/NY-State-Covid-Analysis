**Once we have observed trend of COVID Cases,Deaths and Observed the response of people of US getting tested. We have certain questions "If someone is infected by Covid how likely will they die**

**Let us answer these questions using various rations**

**CFR**
The Case Fatality Rate = the number of confirmed deaths divided by the number of confirmed cases.
It measures the severity of a particular disease by defining the total number of deaths as a proportion of reported cases of a specific disease at a specific time.



```python
df_CFR = df_US.filter(['new_cases_smoothed_per_million','new_deaths_smoothed_per_million','date'], axis=1)
df_CFR['CFR'] = ((df_CFR['new_deaths_smoothed_per_million']/df_CFR['new_cases_smoothed_per_million'])*100)
```

```python
df_CFR = df_CFR.dropna()
df_CFR
```

```python
plt.figure(figsize=(20,4))
plt.plot(df_CFR['date'], df_CFR['CFR'])
plt.xlabel("DATE", fontsize=14)
plt.ylabel("CASE FATALITY RATE", fontsize=14)
#plt.yticks(range(0, 1700000, 200000))
plt.xlabel("DATE",fontsize=14)
plt.suptitle('COVID CASE FATALITY RATE', fontsize=16)
```

**Let us calculate CMR - Crude Mortality Rate**

CMR is defined as  Number of deaths over a given period divided by the person-years lived by the population over that period. It is expressed as number of deaths per 1,000 population.

```python
#Crude Mortality Rate / Crude Death rate 
df_cmr = df_US.filter(['new_deaths_smoothed','population','date'], axis=1)
df_cmr = df_cmr.dropna()
df_cmr['CMR'] = (df_cmr['new_deaths_smoothed']/df_cmr['population'])*100


#Crude Mortality Rate Plot
plt.figure(figsize=(20,4))
plt.plot(df_cmr['date'], df_cmr['CMR'])
plt.xlabel("DATE", fontsize=14)
plt.ylabel("CASE MORATALITY RATE", fontsize=14)
#plt.yticks(range(0, 1700000, 200000))
plt.xlabel("DATE",fontsize=14)
plt.suptitle('COVID CASE MORTALITY RATE', fontsize=16)
```

**IFR - Risk of Death per infection**
Infection Fatality Rate 
if someone is infected with the disease how likely is it that they will die from it?
#IFR = (number of deaths/total number of cases ) * 100

```python
#@title Default title text
df_ifr = df_US.filter(['new_deaths_smoothed','new_cases_smoothed','date'], axis=1)
df_ifr = df_ifr.dropna()
df_ifr['IFR'] = (df_ifr['new_deaths_smoothed']/df_ifr['new_cases_smoothed'])*100


#Infection Fatality Rate Plot
plt.figure(figsize=(20,4))
plt.plot(df_ifr['date'], df_ifr['IFR'])
plt.xlabel("DATE", fontsize=14)
plt.ylabel("INFECTION FATALITY RATE", fontsize=14)
#plt.yticks(range(0, 1700000, 200000))
plt.xlabel("DATE",fontsize=14)
plt.suptitle('COVID INFECTION FATALITY RATE', fontsize=16)
```
